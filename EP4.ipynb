```python
#consulta_cand_2020_AC, escolaridade

a = ['EMC', 'S.C', 'EFC', 'EFI', 'L.E', 'EMI', 'S.I']
y = [1032, 775, 411, 347, 129, 126, 82]

x = np.arange(len(a))

fig, ax = plt.subplots()
rects = ax.bar(x, y)

ax.set_ylabel('Número de candidatos')
ax.set_title('Nível de Instrução')
ax.set_xticks(x)
ax.set_xticklabels(a)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom')



autolabel(rects)
```
